# FILTERING AND ORDERNIG DATAFRAME IN PANDAS

### lets start by importing our dataset

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\ADMIN\Downloads\Pandas files\world_population.csv')

df
```

### Filtering based on the Column 

```python
df[df['Rank'] < 10]
```

```python
df[df['Rank'] <= 10]
```

```python
specific_countries = ['Bangladesh','Brazil']

df[df['Country'].isin(specific_countries)]
```

```python
specific_countries = ['Bangladesh','Brazil','Rusia','Nigeria','China']

df[df['Country'].isin(specific_countries)]
```

```python
df[df['Country'].str.contains('United')]
```

### Filtering based on Index


We will see how to set an index, specifying and axis, using like, 

Loc ----> Location

iloc ------> integer location

```python
##### we will start by setting our index ie country
```

```python
df2 = df.set_index('Country')

df2
```

```python
###### therefore we will use filtering function and specifing the axis you want and also using like
```

```python
df2.filter(items = ['Continent','CCA3'])
```

```python
df2.filter(items = ['Continent','CCA3'], axis = 1)
```

```python
df2.filter(items = ['Continent','CCA3'], axis = 0)
```

```python
df2.filter(items = ['Zimbabwe'], axis = 0)
```

```python
df2.filter(like = 'United', axis = 0)
```

```python
##### Using loc ( looks for the actual value or specific value) and iloc( looks for an integer)
```

```python
df2.loc['United States']
```

```python
df2.iloc[3]
```

# ORDER BY IN PANDAS


Here we can sort our dataframe in ascending, descending, multiple column, single column. 

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\ADMIN\Downloads\Pandas files\world_population.csv')

df
```

```python

df[df['Rank'] < 10]
```

```python

df[df['Rank'] < 10].sort_values(by = 'Rank')
```

```python

df[df['Rank'] < 10].sort_values(by = 'Rank',ascending=[True])
```

```python

df[df['Rank'] < 10].sort_values(by = 'Rank',ascending=[False])
```

```python

df[df['Rank'] < 10].sort_values(by = ['Rank','Country'],ascending=True)
```

```python
df[df['Rank'] < 10].sort_values(by = ['Rank','Country'],ascending=False)
```

```python
df[df['Rank'] < 10].sort_values(by = ['Country','Rank'],ascending=True)
```

```python
df[df['Rank'] < 10].sort_values(by = ['Country','Rank'],ascending=False)
```

```python
df[df['Rank'] < 10].sort_values(by = ['Continent','Country'],ascending=True)
```

```python
df[df['Rank'] < 10].sort_values(by = ['Continent','Country'],ascending=False)
```

```python

df[df['Rank'] < 10].sort_values(by=['Continent','Country'],ascending=[False,True])
```
